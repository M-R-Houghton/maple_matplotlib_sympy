## contourplot/gradplot/streamgradplot

Functions to imitate the existing functions in Maple. Written with the purpose of being more accessible to those who have little or no experience with matplotlib and minimal experience with Python.

All plotting is done using matplotlib and numpy, but to increase ease of use the function $\verb|eqn_to_mesh|$ needs to be capable of interpreting SymPy symbols.

Set up notebook and $\verb|eqn_to_mesh|$ function:

```python
import matplotlib as mpl
import numpy as np
from sympy import *
%matplotlib inline
init_printing()

x,y = symbols('x y')

def eqn_to_mesh(eqn,X,Y):
    '''For substituting SymPy symbols x,y into Numpy meshgrids X,Y.'''
    
    str_eqn = str(eqn)
    sub_eqn = ''
    
    count = 0
    for i in str_eqn:
        
        # Change case of x and y:
        if i == 'x':
            sub_eqn += 'X'
        elif i == 'y':
            sub_eqn += 'Y'
        
        # Change trig functions to Numpy trig functions:
        elif i == 's' and str_eqn[count+1] == 'i':
            sub_eqn += 'np.s'
        elif i == 'c' and str_eqn[count+1] == 'o':
            sub_eqn += 'np.c'
        elif i == 't' and str_eqn[count+1] == 'a':
            sub_eqn += 'np.t'
        elif i == 's' and str_eqn[count+1] == 'q':
            sub_eqn += 'np.s'
        
        # Keep all other characters the same:  
        else:
            sub_eqn += i
        count += 1
            
    return eval(sub_eqn)
```

Declare plotting functions:

```python
def contourplot(f,x_range,y_range,delta=0.01):
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)

    Z = eqn_to_mesh(f,X,Y)
    
    mpl.pyplot.contour(X,Y,Z)
    
    return None


def gradplot(f,x_range,y_range,delta=0.5,colour='r'):
    
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)
    
    Z = eqn_to_mesh(f,X,Y)
    U,V = np.gradient(Z)
    
    mpl.pyplot.quiver(X,Y,U,V,color=colour)
    
    return None


def streamgradplot(f,x_range,y_range,delta=0.5,colour='r'):
    
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)
    
    Z = eqn_to_mesh(f,X,Y)
    U,V = np.gradient(Z)
    
    mpl.pyplot.streamplot(X,Y,U,V,color=colour)
    
    return None
```

Testing $\verb|contourplot|$:

```python
contourplot(x**2 - y**2, [-5,5],[-5,5])
```

Testing $\verb|gradplot|$:

```python
gradplot(x**2+y**2,[-4.0,3.0],[-2.0,2.0])
```

Testing $\verb|streamplot|$:

```python
streamgradplot(x**2 + y**2, [-3,4],[-2,2])
```

Testing multiple plots:

```python
contourplot(x**2 - y**2, [-5,5],[-5,5])
streamgradplot(x**2 - y**2, [-5,5],[-5,5])
```

Testing with trigonometric functions:

```python
contourplot(sin(x)+cos(y),[-5.0,5.0],[-5.0,5.0])
```

```python
contourplot(cos(x**2+y**2)+sin(x)+tan(y),[-5.0,5.0],[-5.0,5.0])
```
