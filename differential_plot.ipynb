```python
import matplotlib as mpl
import numpy as np
from sympy import *
from numpy import pi     # override SymPy object "pi"
%matplotlib inline
init_printing()

x,y = symbols('x y')

def eqn_to_mesh(eqn,X,Y):
    '''For substituting SymPy symbols x,y into Numpy meshgrids X,Y.'''
    
    str_eqn = str(eqn)
    sub_eqn = ''
    
    count = 0
    for i in str_eqn:
        
        # Change case of x and y:
        if i == 'x':
            sub_eqn += 'X'
        elif i == 'y':
            sub_eqn += 'Y'
        
        # Change trig functions to Numpy trig functions:
        elif i == 's' and str_eqn[count+1] == 'i':
            sub_eqn += 'np.s'
        elif i == 'c' and str_eqn[count+1] == 'o':
            sub_eqn += 'np.c'
        elif i == 't' and str_eqn[count+1] == 'a':
            sub_eqn += 'np.t'
            
        # Change other functions to Numpy functions:
        elif i == 's' and str_eqn[count+1] == 'q':
            sub_eqn += 'np.s'
        
        # Keep all other characters the same:  
        else:
            sub_eqn += i
        count += 1
        
    return eval(sub_eqn)
```

```python
## Background code ##

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from numpy import pi

x,y,z = symbols('x y z')

def plot3d(f,xy_range,samples=30):
    xy_lower,xy_upper = xy_range[0],xy_range[1]
    
    lin = np.linspace(xy_lower,xy_upper,samples)
    sam = np.ones(samples)
    
    X = np.outer(lin,sam)
    Y = X.copy().T
    Z = eqn_to_mesh(f,X,Y)
    
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, Z)
    
    return None
```

```python
def gradplot(f,x_range,y_range,delta=0.5,colour='r'):
    
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)
    
    Z = eqn_to_mesh(f,X,Y)
    U,V = np.gradient(Z)
    
    mpl.pyplot.quiver(X,Y,U,V,color=colour)
    
    return None


def streamgradplot(f,x_range,y_range,delta=0.5,colour='r'):
    
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)
    
    Z = eqn_to_mesh(f,X,Y)
    U,V = np.gradient(Z)
    
    mpl.pyplot.streamplot(X,Y,U,V,color=colour)
    
    return None
```

```python
def diffplot(f,x_range,y_range,delta=0.1,colour='r'):
    
    x_lower,x_upper = x_range[0],x_range[1]
    y_lower,y_upper = y_range[0],y_range[1]
    x_vals = np.arange(x_lower,x_upper,delta)
    y_vals = np.arange(y_lower,y_upper,delta)
    X,Y = np.meshgrid(x_vals,y_vals)  

    Z = eqn_to_mesh(f,X,Y)
    #U,V = np.gradient(Z)
    #print Z
    
    X,Y = np.mgrid[x_lower:x_upper:delta, y_lower:y_upper:delta]
    U = np.ones(np.shape(X))
    V = 3*Y-4*Y**3
    
    mpl.pyplot.quiver(X,Y,U,V,color=colour)
    
    return None
```
